# 2026 Data

## Import Previous Data

```python
import pandas as pd

university_df_2016_2025 = pd.read_csv("files/THE World University Rankings 2016-2025 [Processed].csv", thousands=",")
university_df_2016_2025
```

```python
university_df_2016_2025.info()
```

## Import Current Data

### Rankings Tab Data

```python
rankings_2026_df = pd.read_csv('files/university_rankings_2026_raw_rankings.csv')
rankings_2026_df
```

```python
rankings_2026_df.info()
```

### Statistics Tab Data

```python
statistics_2026_df = pd.read_csv('files/university_rankings_2026_raw_statistics.csv')
statistics_2026_df
```

```python
statistics_2026_df.info()
```

## Merge Data From the Two Tabs

```python
combined_2026_df = statistics_2026_df.merge(rankings_2026_df, on=['Rank', 'Name', 'Country'])
combined_2026_df
```

```python
combined_2026_df.info()
```

```python
# Rename 'Industry Score' column to 'Industry Impact'

combined_2026_df = combined_2026_df.rename(columns={"Industry Score": "Industry Impact"})
combined_2026_df.info()
```

```python
# Drop rows with "Reporter" rank status

reporter_index = combined_2026_df.loc[combined_2026_df["Rank"] == "Reporter"].index
combined_2026_df.drop(index=reporter_index, inplace=True)
combined_2026_df.tail()
```

## Calculate Overall Scores

```python
# Use the published formula:
# Overall Score = 0.295(Teaching) + 0.29(Research Environment) + 0.3(Research Quality) + 0.075(International Outlook) +
#                 0.04(Industry Impact)

overall_score = (0.295*combined_2026_df["Teaching"] + 0.29*combined_2026_df["Research Environment"] +
                 0.3*combined_2026_df["Research Quality"] + 0.075*combined_2026_df["International Outlook"] +
                 0.04*combined_2026_df["Industry Impact"]
                )
overall_score
```

```python
combined_2026_df["Overall Score"] = overall_score
combined_2026_df.tail()
```

## Compute University Rankings and Perform Sorting

```python
combined_2026_df["Rank"] = combined_2026_df["Overall Score"].rank(method='min', ascending=False)
combined_2026_df = combined_2026_df.sort_values(by=["Rank"], ignore_index=True)
combined_2026_df
```

```python
# Add 'Year' column

combined_2026_df["Year"] = 2026
combined_2026_df
```

## Combine Previous and Current Data

```python
university_rankings_complete_df = pd.concat([university_df_2016_2025, combined_2026_df])
university_rankings_complete_df
```

```python
# Export as csv
university_rankings_complete_df.to_csv("files/THE World University Rankings 2016-2026 [Processed].csv", index=False)
```
